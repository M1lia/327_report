```python
page_link = 'https://store.steampowered.com/stats/?l=russian'
```

```python
response = requests.get(page_link, headers={'User-Agent': UserAgent().chrome})
response
```

```python
html = response.content
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
table = soup.findAll("table")[1]
table
```

```python
data = []
for tr in table.findAll('tr'):
    data.append([td.get_text() for td in tr.find_all("td")])
```

```python
data
```

```python
data = data[2:]
data = np.array(data)
```

```python
delete_breaks = np.vectorize(lambda s: s.replace('\n', '',  ))

data = delete_breaks(data)
data
```

```python
type(data[0])
```

```python
df_games = pd.DataFrame(data, columns =['Current players', 'Max players today','4delete', 'Game'])
df_games
```

```python
df_games.drop('4delete', axis = 1, inplace = True)
```

```python
df_games.head(5)
```

```python
df_games.info()
```

```python
df_games['Current players'] = df_games['Current players'].apply(lambda s: s.replace(',',''))
```

```python
df_games['Current players'] = df_games['Current players'].astype(int)
```

```python
df_games['Max players today'] = df_games['Max players today'].apply(lambda s: s.replace(',',''))
df_games['Max players today'] = df_games['Max players today'].astype(int)
```

```python
df_games['Game'] = df_games['Game'].astype('string')
```

```python
df_games.info()
```

```python
df_games.to_csv('out.csv')
```
